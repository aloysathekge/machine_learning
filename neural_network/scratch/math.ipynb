## Derivative 

```python

from typing import Callable
import numpy as np
from numpy import ndarray

def deriv(func:Callable[ndarray,ndarray],input_:ndarray, delta:float=0.001)->ndarray:
    """
    Evaluates the derivative of a function func at every element in input_ array

    """
    return (func(input_ + delta) - func(input_-delta))/(2*delta)

def f(x):
   return x**2 + 3*x

input=np.array([2,6,8])

gradients=deriv(f,input)

print(gradients)
```

```python
def matmul_forward(X: ndarray,
    W: ndarray) -> ndarray:
    '''
    Computes the forward pass of a matrix multiplication.
    '''
    assert X.shape[1] == W.shape[0], \
    '''
    For matrix multiplication, the number of columns in the first array should
    match the number of rows in the second; instead the number of columns in the
    first array is {0} and the number of rows in the second array is {1}.
    '''.format(X.shape[1], W.shape[0])
    # matrix multiplication
    N = np.dot(X, W)
    return N

x=np.array([[1,3,4,5],[1,3,4,5]])
w=np.array([[0.1],[1.3],[0.4],[1.5]])
matmul_forward(x,w)
```
