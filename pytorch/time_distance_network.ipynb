```python
import torch
import torch.nn as nn
import torch.optim as optim
import helper_utils

torch.manual_seed(42)
```

```python
# Distances in miles for recent bike deliveries
distances = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)

# Corresponding delivery times in minutes
times = torch.tensor([[6.96], [12.11], [16.77], [22.21]], dtype=torch.float32)
```

```python
model=nn.Sequential(
    nn.Linear(1,3),
    nn.ReLU(),
    nn.Linear(3,1)
)
```

```python
loss_function=nn.MSELoss()
optimizer=optim.SGD(model.parameters(),lr=0.01)
```

```python
for epoch in range(500):
    #Training loop
    optimizer.zero_grad()

    predictions=model(distances)

    loss=loss_function(predictions,times)

    loss.backward()

    optimizer.step()
    print(f"for {epoch} loss is : {loss.item():.4f}")
```

```python
helper_utils.plot_results(model, distances, times)
```

```python
distance_to_predict = 7.0
```

```python
with torch.no_grad():
    new_distance=torch.tensor([[distance_to_predict]],dtype=torch.float32)

    output=model(new_distance)

    print(f"predicted output for {distance_to_predict} is {output.item():.1f}")
```
